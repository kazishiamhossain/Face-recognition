```python
plt.subplot(3,1,1)
plt.imshow(r,cmap='gray')
plt.subplot(3,1,2)
plt.imshow(g,cmap='gray')
plt.subplot(3,1,3)
plt.imshow(b,cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,3,1)
plt.title('Red region')
plt.imshow(r,cmap='gray')
plt.subplot(1,3,2)
plt.title('Green region')
plt.imshow(g,cmap='gray')
plt.subplot(1,3,3)
plt.title('Blue region')
plt.imshow(b,cmap='gray')
plt.show()
```
